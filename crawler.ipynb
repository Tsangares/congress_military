```python
import os
from bs4 import BeautifulSoup
import requests,json,time,random
```

```python
with open('cred.json') as f:
    cred = json.load(f)
```

```python
session = requests.Session()
session.get("https://library.cqpress.com/congress/login.php")
data = {
    'requested': '/congress/index.php',
    'username': cred['username'],
    'password': cred['password'],
    'login': 'Login',
}
login = session.post('https://library.cqpress.com/congress/login.php', data=data)
with open("html/login_response.html","w+") as f:
    f.write(login.html)
```

```python
session.headers['User-Agent'] =  'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3626.28 Safari/537.36'
session.headers['Accept'] = 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,*/*;q=0.8'
```

```python
url = "https://library.cqpress.com/congress/memberanalysisresults.php?firstname=&lastname=&congress1=&posyear1=&position1=&congress2=&posyear2=&position2=&party=&state=&yearlimit=0&year1=&year2=&religion=&race=&education=&occupation=&gender=&milservice="
directory = session.get(url)
with open("html/directory.html","w+") as f:
    f.write(directory.text)
```

```python
soup = BeautifulSoup(directory.text,'html')
```

```python
directory = []
for i,row in enumerate(soup.find_all('tr')):
    if 'valign' in row.attrs or 'bgcolor' in row.attrs: continue #header
    person = {}
    for j,col in enumerate(row.find_all('td')):
        
        if j==0:
            #Button
            person['id'] = col.find('input').attrs['value']
        elif j==1:
            #name
            person['link'] = col.find('a').attrs['href']
            person['name'] = col.text.strip()
        elif j==2:
            #Dem
            person['party'] = col.text
        elif j==3:
            person['state'] = col.text
        elif j==4:
            person['affilication'] = col.text
        elif j==5:
            person['term'] = col.text
    assert(len(person.keys())==7)
    directory.append(person)
with open('dist/directory.json','w+') as f:
    json.dump(directory,f,indent=2)
    
```

```python
base = "https://library.cqpress.com/congress/"
for i,person in enumerate(directory):
    url = base + person['link']
    detail = session.get(url)
    print(detail)
    with open(f'people_html/{person["id"]}.html','w+') as f:
        f.write(detail.text)
        print(f"Wrote {i}:",f'people_html/{person["id"]}.html')
    time.sleep(random.random())
```
